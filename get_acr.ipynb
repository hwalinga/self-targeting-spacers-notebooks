```python
import tabula
import pandas as pd 
import numpy as np 
acr_pub = '/home/hielke/bep/misc/fnz098.pdf'
target_folder = '/home/hielke/bep/acr/'
acr_range = '25-32'
aca_range = '33'
```

```python
pd.set_option('display.max_rows', 500)
```

```python
acr_df = tabula.read_pdf(acr_pub, pages=acr_range)
acr_df
```

```python
# Remove repeated header
acr_df = acr_df.drop(index=acr_df.query('(Family == "Fami" | Family == "ly") | (Accession == "numb" | Accession == "er")').index)
acr_df = acr_df.drop(index=range(0, 4))
import itertools as it 
import operator as op
acr_pdf
```

```python
groups = it.groupby(zip(acr_df['Accession'].values, acr_df['Size'].values), key=op.itemgetter(1))
with open(target_folder + 'acr_ids', 'w') as acr_file:
    for k, g in groups: 
        g = list(g)
        joined = it.chain(g, next(groups)[1])
        acc = ''.join(it.filterfalse(pd.isnull, map(op.itemgetter(0), joined)))
        print(acc, file=acr_file)
```

```python
aca_df = tabula.read_pdf(acr_pub, pages=aca_range)
aca_df
```

```python
with open(target_folder + 'aca_ids', 'w') as aca_file:
    for aca_id in it.filterfalse(pd.isnull, aca_df['Accession number'].values):
        print(aca_id, file=aca_file)
```
